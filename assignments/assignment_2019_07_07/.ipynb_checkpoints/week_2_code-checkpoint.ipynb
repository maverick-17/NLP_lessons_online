```python
import json
import requests
from collections import defaultdict
```

```python
url = 'http://map.amap.com/subway/index.html?&1100'

guangzhou_metro_url = 'http://map.amap.com/service/subway?_1555502190153&srhdata=4401_drw_guangzhou.json'
```

```python
response = requests.get(guangzhou_metro_url)
gz_json = json.loads(response.text)
info = gz_json.get('l')
```

```python
def get_gz_metro_station_info(url):
    response = requests.get(url)
    gz_json = json.loads(response.text)
    info = gz_json.get('l')
    
    result = []
    for line in info:
        line_dict = defaultdict(list)
        line_name = line['ln']
        for st in line['st']:
            item = {}
            station = st['n']
            postions = tuple(float(k) for k in st['sl'].split(','))
            #         print(station, postions)
            item[station] = postions

            line_dict[line_name].append(item)
        print(line_dict)
        result.append(line_dict)

    return result
```

```python
gz_metro_info = get_gz_metro_station_info(guangzhou_metro_url)
```

```python
gz_all_st_pos = {}
for item in gz_metro_info:
    for line in item:
#         print(item[line])
        for st in item[line]:
            for st_name, pos in st.items(): 
                if st_name in gz_all_st_pos: continue
                gz_all_st_pos[st_name] = pos
print(gz_all_st_pos)
```

```python
len(gz_all_st_pos)
```

```python
gz_metro_line_info = []
for item in gz_metro_info:
    line_connect = []
    for line in item:
#         print(item[line])
        for st in item[line]:
            line_connect.append(list(st.keys())[0])
        gz_metro_line_info.append(line_connect)    
print(gz_metro_line_info)
```

```python
gz_all_st_connection = defaultdict(list)
for line in gz_metro_line_info:
    for i, st in enumerate(line):
        if st == line[0]:
            gz_all_st_connection[st].append(line[1])
        if st == line[-1]:
            gz_all_st_connection[st].append(line[-2])
        else:
            gz_all_st_connection[st].extend([line[i-1], line[i+1]])

print(gz_all_st_connection)
```

```python
metro_graph.add_nodes_from(list(gz_all_st_pos.keys()))
```

```python
nx.draw(metro_graph, gz_all_st_pos, with_labels=True, node_size=30)
```

```python
def search_metro_path(start, end, graph_connection):
    if start not in gz_all_st_pos or end not in gz_all_st_pos:
        raise Exception('invalidate gz metro station, try again')
    
    pathes = [[start]]
    visited = set()
    
    while pathes:
        path = pathes.pop(0)
        frontier = path[-1]
        
        successors = graph_connection[frontier]
        if frontier in visited:
            continue
            
        for succesor in successors:
            new_path = path + [succesor] # DFS
            
           new_path = [succesor] + path # WFS
            
            pathes.append(new_path)
            
            if succesor == end:
                return new_path
        
        visited.add(frontier)
    
```

```python
search_metro_path('广州南站', '客村', gz_all_st_connection)
```

## loss

```python
def loss(y, y_hat):
    return sum((y_i - y_hat_i) ** 2 for y_i, y_hat_i in zip(list(y), list(y_hat))) /  len(list(y))
                                                           
                                                           
```

### Method_1 random generate best y &  b

```python
def try_random_generate_line(n=1000):
    min_loss = float('inf')
    
    for i in range(n):
        k = random.randint(-100, 100)
        b = random.randint(-100, 100)
        price_hat = [price(rm, k, b) for rm in X_rm]
        
        current_loss = loss(y, price_hat)
        
        if current_loss < min_loss:
            min_loss = current_loss
            best_k, best_b = k, b
            print(f'In time {i}, loss is {min_loss} get better k: {best_k}, get better b: {best_b}')       
        
```

```python
try_random_generate_line(2000)
```

### Method 2 Direction adjustment

```python
direction = [
    (+1, -1),
    (-1, -1),
    (-1, -1),
    (+1, -1) 
]

# next_direction = random.choice(direction)
step = 1

def try_generate_line_with_direction(n=1000):
    next_direction = random.choice(direction)
    best_k = random.randint(-100, 100)
    best_b = random.randint(-100, 100)
    min_loss = float('inf')  
    
    for i in range(n):
        k_direction, b_direction = next_direction
        current_k = best_k + k_direction * step
        current_b = best_b + b_direction * step
                
        price_hat = [price(rm, current_k, current_b) for rm in X_rm]
        
        current_loss = loss(y, price_hat)
        
#         print(f'current k: {current_k}, current_b : {current_b}, current_loss: {current_loss}')
  
        if current_loss < min_loss:
            min_loss = current_loss
            best_k, best_b = current_k, current_b
            next_direction = next_direction
            print(f'In time {i}, loss is {min_loss} get better k: {best_k}, get better b: {best_b}')
            print(f'time {i} current_direction is {next_direction}')
        else:
#             next_direction = random.choice([d for d in direction if d != next_direction])
              next_direction = random.choice(direction)
```

```python
try_generate_line_with_direction(2000)
```
